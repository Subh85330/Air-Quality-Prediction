```python
df = df2.merge(df1, on='Date_Time')
df
```

```python
DF = df.iloc[:, 2:]
DF
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(DF)
cols = DF.columns
DF_normalized = pd.DataFrame(np_scaled, columns = cols)
DF_normalized
```

```python
FINAL = DF_normalized.iloc[:10000]
FINAL
```

```python
x = FINAL[['pressure_set_1','air_temp_set_1','relative_humidity_set_1','wind_speed_set_1','wind_direction_set_1','dew_point_temperature_set_1d','Latitude','Longitude']]
x
```

```python
y = FINAL[['Sample Measurement']]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.4,random_state = 101)
```

```python
x_train
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
predictions = lm.predict(x_test)
predictions
```
